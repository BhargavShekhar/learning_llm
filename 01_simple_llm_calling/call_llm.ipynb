```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemma-4-31b-it"
)

response = llm.invoke("How many moon does jupitor have?")

print(response.text)
```

```python
response = llm.invoke([
    SystemMessage(content="You are a helpful assistant that answers in one short line."),
    HumanMessage(content="How many moons does Jupiter have?")
])
print(response.text)
```

```python
prompt = "Wrtie a single tag line for a quirky coffe shop. Only output the tag line nothing else."

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature = 0
)
response = llm.invoke(prompt)
print(response.text)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature = 1
)
response = llm.invoke(prompt)
print(response.text)
```

```python
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
llm = ChatGroq(
    model="qwen/qwen3-32b",
    reasoning_format="parsed",
)

messages = [
    SystemMessage(
        content="You are a helpful assistant that explains concepts briefly and clearly."
    ),
    HumanMessage(
        content="Explain what an LLM is in 2 simple sentences."
    )
]

response = llm.invoke(messages)

print(response.content)
```
