```python
import chromadb

client = chromadb.Client()
```

```python
collection = client.create_collection("news_article")
```

```python
collection.add(
    ids = ["id1", "id2", "id3", "id4", "id5"],
    documents = [
        "India won the cricket match yesterday.",
        "OpenAI released a new AI model.",
        "Heavy rain caused flooding in Mumbai.",
        "Tesla launched a new electric car.",
        "SpaceX got NASA contracts worth 8 billion dollars.",
    ]
)
```

```python
data = collection.get(
    include=["documents", "embeddings", "metadatas"]
)

# print embedding
for i, emb in enumerate(data["embeddings"]):
    print(f"\nDocument {i+1}")
    print(f"Text: {data['documents'][i]}")
    print("Embedding length:", len(emb))
    print("First 10 values:", emb[:10])
```

```python
results = collection.query(
    query_texts = ["This is a query related to Elon Musk"],
    n_results = 2
)

print(results)
```
